##### Selecting columns of interest to test the hypothesis

##### Creating a dataframe with participants' self-assessment scores at Time 1

```python
self_assessment_1 = ["attr3_1","intel3_1","sinc3_1","fun3_1","amb3_1", "attr5_1","intel5_1","sinc5_1","fun5_1","amb5_1"]
data_1 = df[self_assessment_1]
data_1.shape
```

##### Creating a dataframe with participants' self-assessment scores at Time 3

```python
self_assessment_2 = ["attr5_3","intel5_3","sinc5_3","fun5_3","amb5_3", "attr3_3","intel3_3","sinc3_3","fun3_3","amb3_3"]
data_2 = df[self_assessment_2]
data_2.shape
```

##### Concatenating the two dataframes to obtain participants' self-assessment scores at Time 1 and Time 3

```python
data_ok= pd.concat([data_1, data_2], axis=1).fillna(0)
```

```python
data_ok.shape
```

```python
data_ok
```

```python
data_ok.shape
```

##### Adding a new column "Sign up" in the dataframe with participants' summed self-assessment scores at Time 1

```python
data_ok["Sign_up"] = data_ok["attr3_1"]+data_ok["sinc3_1"]+data_ok["intel3_1"]+data_ok["fun3_1"]+data_ok["amb3_1"]+data_ok["attr5_1"]+data_ok["sinc5_1"]+data_ok["intel5_1"]+data_ok["fun5_1"]+data_ok["amb5_1"]
```

```python
data_ok.head()
```

##### Adding a new column "After" in the dataframe with participants' summed self-assessment scores at Time 3

```python
data_ok["After"] = data_ok["attr3_3"]+data_ok["sinc3_3"]+data_ok["intel3_3"]+data_ok["fun3_3"]+data_ok["amb3_3"]+data_ok["attr5_3"]+data_ok["sinc5_3"]+data_ok["intel5_3"]+data_ok["fun5_3"]+data_ok["amb5_3"]
```

```python
data_ok.head()
```

##### Adding a new column "Delta" in the dataframe with participants' difference self-assessment scores at Time 1 and Time 3

```python
data_ok["Delta"] = data_ok["After"]-data_ok["Sign_up"]
```

```python
data_ok.head()
```

##### Finding the unique values of the diffrence in scores between Time 1 and Time 3

```python
data_ok["Delta"].nunique()
```

##### Listing the unique values of the diffrence in scores between Time 1 and Time 3

```python
data_ok["Delta"].unique()
```

```python
diff = data_ok["Delta"].unique()
```

##### Sorting the unique values of the diffrence in scores between Time 1 and Time 3 in descending order

```python
#Sort the array in descending order    
for i in range(0, len(diff)):    
    for j in range(i+1, len(diff)):    
        if(diff[i] < diff[j]):    
            temp = diff[i];    
            diff[i] = diff[j];    
            diff[j] = temp;    
     
print();    
     
#Displaying elements of the array after sorting    
    
print("Elements of array sorted in descending order: ");    
for i in range(0, len(diff)):    
    print(diff[i], end=" ");
```

##### Visualizing the difference between self assessment at sign up and self assessment 3-4 weeks after the speed dating session

##### Creating a dataframe with the Sign_up After and Delta columns

```python
compare_df = data_ok[["Sign_up", "After", "Delta"]]
compare_df.head()
```

##### Concatenating this dataframe with the original df dataframe

```python
data_complete = pd.concat([df, compare_df], axis=1)
```

```python
data_complete.head()
```

##### Grouping the iid and Delta to have the diffrence between Time 1 and Time 3 for each participant

```python
grouped_df = data_complete.groupby(["iid", "Delta"], as_index=False)
```

```python
grouped_df.first()
```

```python
import numpy as np
my_df = grouped_df.aggregate(np.sum)
```

```python
my_df
```

##### Plotting a graph to show the difference between scores at Time 1 and Time 3 for each participant

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x = "iid", y = "Delta", data = my_df,
            palette = "hls",
            capsize = 0.05,             
            saturation = 8
           )
```
